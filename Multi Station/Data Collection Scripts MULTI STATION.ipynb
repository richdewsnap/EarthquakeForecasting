### EMAIL GENERATOR CODE

Code takes the date-time string from the Wilber 3 website for each earthquake and creates the email format necessary to retrieve the data



```python
string = "AAAAA"

timewindow = timedelta(hours=10)

outString =defaultString.replace("////NAME////",NAME)

while string.lower() != "end":

  for station in StationNames:

    print("\n"+station)
    string = input("Copy and paste the date-time string for station "+station) 
    #FORMAT: 2022-05-22 15:18:35
    
    if string.lower()=="end":
      print("\n\nENDING")
      break

    string = string.strip()
    print(string)
    chunks = string.split(" ")

    #print(chunks[0])
    #print(chunks[1])
    #print(chunks[2])

    dateinfo = chunks[0].split("-")
    timeinfo = chunks[1].split(":")

    Pwave = datetime(int(dateinfo[0]), int(dateinfo[1]), int(dateinfo[2]), int(timeinfo[0]), int(timeinfo[1]), int(timeinfo[2]), 000)

    start = Pwave - timewindow

    print(start)
    print(Pwave)

    tempString = "\n"+station + " " + str(start) + ".000 " + str(Pwave) +".000 "+ EndString
    tempString = tempString.replace("-"," ")
    tempString = tempString.replace(":"," ")
    print(tempString)

    outString+=tempString

  if string.lower()!="end":
    print()
    print()
    print(outString)



#FORMAT: 2022-05-22 15:18:35
```
